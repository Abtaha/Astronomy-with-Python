# Astronomy with Python using AstroPy

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "data.fits"

hdu_list = fits.open(file)

image = hdu_list[0].data

for i in range(4):
    plt.imshow(image[i], cmap="gray")
    plt.colorbar()
    plt.show()
```

## Image Data

Mainly Stars. Not that high brightness. Number of stars is also quite low. Average temperature is also very low. Maximum temperature is very high though.

```python
print('Min:', np.min(image))
print('Max:', np.max(image))
print('Average:', np.mean(image))
```
